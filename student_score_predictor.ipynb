```python
import pandas as pd

url = "http://bit.ly/w-data"
data = pd.read_csv(url)

print("Data preview:")
print(data.head())

import matplotlib.pyplot as plt

plt.scatter(data['Hours'], data['Scores'], color='blue')
plt.title('Hours Studied vs Percentage Score')
plt.xlabel('Hours')
plt.ylabel('Score')
plt.grid(True)
plt.show()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 1: Split data into input (X) and output (y)
X = data[['Hours']]  # Features (input)
y = data['Scores']   # Labels (output)

# Step 2: Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 3: Train the model
model = LinearRegression()
model.fit(X_train, y_train)

print("Model trained ✅")

# Predicting test results
predictions = model.predict(X_test)

# Plotting the regression line on top of the scatter plot
line = model.coef_ * X + model.intercept_

plt.scatter(X, y)
plt.plot(X, line, color='red')  # regression line
plt.title("Regression Line Fit")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.grid(True)
plt.show()

# Compare actual vs predicted
print("Predicted vs Actual Scores:")
for pred, actual in zip(predictions, y_test):
    print(f"Predicted: {pred:.2f}, Actual: {actual}")
```
